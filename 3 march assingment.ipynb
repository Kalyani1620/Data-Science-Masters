#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

* **Heatmap**:-A heatmap is a two-dimensional graphical representation of data where the individual values that are contained in a matrix are represented as colours. 

* **Histogram**:- A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within discrete bins.

* **Bar Plot**:-Seaborn supports many types of bar plots. We combine seaborn with matplotlib to demonstrate several plots. Several data sets are included with seaborn (titanic and others), but this is only a demo. You can pass any type of data to the plots.


* **CatPlot** (Categorical Plot):-Seaborn catplot method enables the work efficiently by using the definite and render data defined in a parameter. The technique produces the object of the facet grid; it is used to plot the graphs for several types of aspects. It utilizes the scatterplot for the datasets.


* **Density Plot**:-A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions.


* **Joint Distribution Plot**:-The joint plot is the concise way of understanding the relationship between two variables as well as the individual distribution of each variable. The Joint plot is consists of 3 separate plots. In which one is the middle figure which is used to see the relationship between x and y.

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions. 
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
fmri = sns.load_dataset("fmri")
sns.lineplot( x = "timepoint",
			y = "signal",
			data = fmri);
```

#### Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.

```python
import pandas as pd 
import seaborn as sns
iris = sns.load_dataset('iris')
my_data_frame = pd.DataFrame(iris)
my_data_frame.head()
```

```python
g = sns.pairplot(my_data_frame)
```

```python
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"], \
                markers=["o", "s", "D"], hue="species")
```

#### Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(0)
sns.set()
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data, vmin=0, vmax=1)
plt.show()
```
